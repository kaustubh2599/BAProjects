```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib 
import openaq
```

```python
api = openaq.OpenAQ()
```

```python
London = api.measurements(city='London', limit=5000, df=True)
Paris = api.measurements(city='Paris', limit=5000, df=True)
Madrid = api.measurements(city='Madrid', limit=5000, df=True)
Berlin = api.measurements(city='Berlin', limit=5000, df=True)
Rome = api.measurements(city='Roma', limit=5000, df=True)
```

### Compiling All Cities in a DataFrame named Cities 

```python
Cities=pd.concat([London,Paris,Madrid, Berlin, Rome],axis=0)
```

```python
Cities
```

### Most Polluted Cities in terms of SO2, CO and pm25 

```python
AirQualityCO = Cities[Cities['parameter']== 'co']
```

```python
AirQualityCO 
```

```python
AirQualityCO.head(6)
```

```python
fig, ax = plt.subplots(1, figsize=(14,10))


ax = sns.boxplot(
    x='city',
    y='value',
    data=AirQualityCO,
    fliersize=0,
    palette='deep',
    ax=ax)


ax.set_ylim([0,1000])
ax.set_ylabel("CO", fontsize=20)
ax.set_xlabel("city")


sns.despine(offset=10)

plt.show()
```

### Roma City is most polluted interms of CO

### Definiing a function which, given a number, computes the square root

```python
def Sqrt_Value (x):
    x= np.sqrt(x)
    return x
x=9
```

```python
Sqrt_Value(x)
```

### Using the function to compute the square root of the column value in the Cities DataFrame

```python
Cities['Sqrt_Value'] = Cities['value'].apply(Sqrt_Value)
```

```python
Cities
```

### Computing the Square Root of the Column Values using a Lambda Function

```python
Cities['Updated_Values']= Cities['value'].apply(lambda x : Sqrt_Value(x))
```
